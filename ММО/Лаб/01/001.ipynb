# Импорт бииблиотек

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Чтение данных

Этот набор данных был сформирован из отзывов о местах отдыха, опубликованных 249 рецензентами на сайте holidayiq.com до октября 2014 года. Рассматривались отзывы, относящиеся к 6 категориям среди направлений Южной Индии, и подсчитывалось количество отзывов в каждой категории для каждого рецензента

```python
df = pd.read_csv('buddymove_holidayiq.csv')
df = df.sort_values(by="Sports")
df.reset_index(inplace=True, drop=True)
df
```

Мы распологаем чсловыми (NUMERIC) типами данных, их несколько (SEVERAL NUMERIC) и они не упорядочены (NOT ORDERED) или слабо упорядочены. Соответственно для этого используем BOXPLOT, `lineplot`, `stackplot`, `correlogram`, `boxplot`, `violinplot`

Отобразим процент отзывов при помощи `pie`. Как видно `Sports` иммеет ощутимо меньш всего отцывов

```python
plt.pie(
    df[["Sports", "Religious", "Nature", "Theatre", "Shopping", "Picnic"]].sum(axis=0), 
    labels=["Sports", "Religious", "Nature", "Theatre", "Shopping", "Picnic"],
    autopct='%.0f%%'
)
plt.legend()
```

Как видно `lineplot` неудачный выбор, тк данные сильно пересекают друг друга 

```python
plt.plot(
    df.index, 
    df[["Sports", "Religious", "Nature", "Theatre", "Shopping", "Picnic"]], 
    label=["Sports", "Religious", "Nature", "Theatre", "Shopping", "Picnic"]
    )
plt.legend()
```

Создадим `stackplot` для представления о  том как выглядят данные. Видна линейная зависимость данных

```python
plt.stackplot(
    df.index, 
    df[["Sports", "Religious", "Nature", "Theatre", "Shopping", "Picnic"]].T, 
    labels=["Sports", "Religious", "Nature", "Theatre", "Shopping", "Picnic"]
    )
plt.legend()
```

```python
plt.stackplot(
    df.index, 
    df[["Sports", "Religious", "Nature", "Theatre", "Shopping", "Picnic"]].rolling(25).mean().T,
    labels=["Sports", "Religious", "Nature", "Theatre", "Shopping", "Picnic"]
    )
plt.legend()
```

Отобразим `boxplot` и `violinplot` для понимания распределения данных

```python
plt.boxplot(
    df[["Sports", "Religious", "Nature", "Theatre", "Shopping", "Picnic"]],
    labels=["Sports", "Religious", "Nature", "Theatre", "Shopping", "Picnic"]
    );
```

```python
plt.violinplot(
    df[["Sports", "Religious", "Nature", "Theatre", "Shopping", "Picnic"]],
    );
plt.xticks(
    np.arange(len(["Sports", "Religious", "Nature", "Theatre", "Shopping", "Picnic"]))+1,
    ["Sports", "Religious", "Nature", "Theatre", "Shopping", "Picnic"]
);
```

Неудачные scatterplot
